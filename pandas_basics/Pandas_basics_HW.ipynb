**Задание 2**

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
```

```python
data.head()
```

```python
data = data[(data['country'].isin(['Lithuania', 'Latvia', 'Estonia'])) & (data['category'].isin([4, 12, 21])) & ((2005 <= data['year']) & ( data['year'] <=2010) ) & (data['quantity'] >= 0)]


```

```python
sum(data.groupby('country').sum()['quantity'].to_list())
```

```python
(2005 <= data['year']) & ( data['year'] <=2010) &  (data['quantity'] >= 0)
```

```python
##
filtered_countries = data[data['country'].isin(['Lithuania', 'Latvia', 'Estonia'])]
                          

filtered_countries.head()
```

```python
filtered_data = filtered_countries[ (2005 <= filtered_countries['year']) & ( filtered_countries['year'] <=2010) ]
filtered_data
```

```python
# filtered_category = filtered_data[ ( filtered_data['category'] == 4) | ( filtered_data['category'] == 12) | ( filtered_data['category'] == 21) ]

filtered_category = filtered_data[filtered_data['category'].isin([4, 12, 21])]
filtered_category

```

```python
filtered_quantity = filtered_category[filtered_category['quantity'] >= 0]
filtered_quantity
```

```python
filtered_quantity.sum()
```

```python
result2 = filtered_quantity.groupby('country').sum()
```

```python
sum(result2['quantity'].to_list())
```
