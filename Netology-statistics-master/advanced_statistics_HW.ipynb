Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 

```python
house_price = data['MEDV']
lower_population = data['LSTAT']
```

```python
house_price = np.array(house_price).reshape(-1,1)
lower_population = np.array(lower_population).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    lower_population,
    house_price,
    test_size = 0.2
)
```

```python
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
```

```python
print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('% lower status of the population')
plt.ylabel('house price')
plt.title('Линейная регрессия от одной переменной')
# Построим прямую для регрессии
# Создадим 100 точек для предсказаний
x = np.linspace(-10, 60, 60)
# Посчитаем предсказанные значения
# y = kx + b
# k = model.coef_
# b = model.intercept_
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(-10,60)
plt.ylim(-10,60)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

В данном случае получается зависимость двух переменных: **чем ниже социальный статус жителей района, тем ниже цена на жилье.**

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
# Количество испытаний
trials = 10

# Массив значений
values = np.random.randint(1, 7, size=trials)
plt.hist(values, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5], edgecolor='white')

plt.xlabel('Значение')
plt.ylabel('Количества раз')
plt.title('Подбрасывание кубика ' + str(trials) + ' раз')
plt.show()
```

**Это равномерное распределение**

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
number_of_sums = 1000
trials = 100
X1 = np.zeros([number_of_sums])
for i in range(number_of_sums):
    # суммируем все элементы равномерных выборок
    X1[i] = np.sum(np.random.randint(1, 7, size=trials))/trials
plt.hist(X1, 50)
plt.xlabel('X values')
plt.ylabel('Number of elements in each X')
plt.show()
```

**Распределение - нормальное.  можно применить центральную предельную теорему, потому что исходное распределение не слишком скошенно. также мы взяли большое кол-во выборки.**
